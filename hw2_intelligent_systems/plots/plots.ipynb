```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.ticker import FuncFormatter
```

```python
sns.set_style('darkgrid')
palette = sns.color_palette('Accent', 10)
# palette = sns.color_palette('Paired', 12)
sns.set_palette(palette)
```

```python
df = pd.read_csv('../data/test2.csv')
# Create the eval_strategy column based on random_size values
df['eval_strategy'] = df['random_size'].map({0.1: "rnd_0.1", 0.2: "rnd_0.2", 1.0: "full"})
df
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
# sns.boxplot(df, x='population_size', y='score', hue='eval_strategy', dodge=True, palette=palette, width=0.5)
sns.boxplot(df, x='population_size', y='score', dodge=True, palette=palette, width=0.5)
ax.axhline(y=65.2, label='Weighted MCTS', color=palette[4], linestyle='--', linewidth=1)
ax.axhline(y=45.7, label='Unweighted MCTS', color=palette[5], linestyle='--', linewidth=1)
plt.legend()
plt.title('Population Size vs. Winrate [%]')
plt.xlabel('Population Size')
plt.ylabel('Winrate [%]')
plt.savefig('population_size.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.boxplot(df, x='eval_strategy', y='score', dodge=True, palette=palette, width=0.5)
ax.axhline(y=65.2, label='Weighted MCTS', color=palette[4], linestyle='--', linewidth=1)
ax.axhline(y=45.7, label='Unweighted MCTS', color=palette[5], linestyle='--', linewidth=1)
plt.legend()
plt.title('Evaluation Size vs. Winrate [%]')
plt.xlabel('Eval Size')
plt.ylabel('Winrate [%]')
plt.savefig('eval_size.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
df = pd.read_csv('../data/det_rnd_eval_test.csv')
df
```

```python
df_copy = df[~df['eval_strategy'].isin(['default_mcts', 'default_mcts_no_weights'])]
fig, ax = plt.subplots(figsize=(8, 8))
# sns.boxplot(df, x='population_size', y='score', hue='eval_strategy', dodge=True, palette=palette, width=0.5)
sns.boxplot(df_copy, x='eval_strategy', y='score', dodge=True, palette=palette, width=0.5)
ax.axhline(y=df[df['eval_strategy'] == 'default_mcts']['score'].iloc[0], label='Weighted MCTS', color=palette[4], linestyle='--', linewidth=1)
ax.axhline(y=df[df['eval_strategy'] == 'default_mcts_no_weights']['score'].iloc[0], label='Unweighted MCTS', color=palette[5], linestyle='--', linewidth=1)
plt.legend()
plt.title('Evaluation strategy vs. Winrate [%]')
plt.xlabel('Evaluation strategy')
plt.ylabel('Winrate [%]')
plt.savefig('evaluation_strategy.png', dpi=300, bbox_inches='tight')
plt.show()
```
